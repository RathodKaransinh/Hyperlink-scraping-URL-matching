# Import libraries 

```python
import requests
from bs4 import BeautifulSoup
```

# Enter Domain link you want to scrap

```python
domain = input()
```

int_links => store links that contain the given domain 
ext_links => store links that not contain the domain 
int_link_t, ext_link_t => store anchor tag text  
broken_links => status code not equal to 200



```python
int_links = []
ext_links = []
int_link_t = []
ext_link_t = []
broken_links = []
```

append the input url and text in 'int_links' and 'int_link_t'

```python
int_links.append(domain)
int_link_t.append('Analytics vidhya') # enter a domain text
```

# Find all anchor tag href and text of particular link
=> use bs4 and requests library 


```python
def hyper_link(url):

    global int_links,ext_links,int_link_t,ext_link_t
    try:
        grab = requests.get(url,headers={'User-Agent': 'Mozilla/5.0'})
        if grab.status_code != 200:
            pass
        else:
            # remove the link and text from the list 
            i = int_links.index(url)
            int_links.remove(url)
            txt = int_link_t.pop(i)
            broken_links.append(url)
            print('broken link removed',url, txt)
            return

        soup = BeautifulSoup(grab.text,'html.parser')

        # for handling relative path 
        global domain 
        
        # find all anchor tag
        for anc_tag in soup.find_all("a"):

            # find 'href' attribute 
            href = anc_tag.get('href')

            # get the text of anchor tag
            text = anc_tag.text.lower().strip()

            # if this link present in our list then we go for the next anchor tag
            if (href in int_links) or (href in ext_links) or (href in broken_links):
                pass
            else:

                if (href != "") and (href != "#") and href != None:

                    # if link have absolute path
                    if href.startswith('http'):
                            if domain in href:
                                int_links.append(href)
                                int_link_t.append(text)
                                print(href,text) 

                            else:
                                ext_links.append(href)
                                ext_link_t.append(text)
                                print('ext',href,text)

                    else:
                        if href.startswith('/'):
                            href = domain + href
                        else:
                            href = domain + '/' + href

                        # if this link present in our list then we go for the next anchor tag
                        if (href in int_links) or (href in ext_links) or (href in broken_links):
                                pass
                        else:
                                if domain in href:
                                    int_links.append(href)
                                    int_link_t.append(text)
                                    print(href,text)
                                else:
                                    ext_links.append(href)
                                    ext_link_t.append(text)
                                    print('ext',href,text)
    except:
        print('Url not valid')
```

# find all hyper link
=> traverse int_links for finding all hyper link

```python
for link in int_links:
    hyper_link(link)
```

# find all Keyword of url 

```python
def keyWord(url):
    ch = [',','.','_','-','/','  ']
    for c in ch:
        url = url.replace(c,' ')
    return url
```

# store the links in text file with corresponding anchor tag text and link keyword

```python
f = open('/content/internal_links.txt','w')
for i in range(len(int_links)):
    sent = keyWord(int_links[i])
    f.write(int_links[i] + ',' + int_link_t[i].lower()+ ' ' + sent.lower())
    f.write("\n")
f.close()
```

```python
f = open('/content/external_links.txt','w')
for i in range(len(ext_links)):
    sent = keyWord(ext_links[i])
    f.write(ext_links[i] + ',' + ext_link_t[i].lower()+ ' ' + sent.lower())
    f.write("\n")
f.close()
```

# URL matching



get the list of url and that text for matching 

```python
f = open("/content/internal_links.txt",'r')
lines = f.readlines()
urls_list = []
urls_keyword = []
for line in lines:
    line = line.split(',',1)
    if len(line) >1:
        urls_list.append(line[0])
        urls_keyword.append(line[1])
print(len(urls_list))
f.close()
```

# 'find_link' function :
=> this will take user query as argument

```python
def find_link(query):
    tokens = query.split(' ')
    global urls_list , urls_keyword
    j = 0
    count = []
    
    for url_keyword in urls_keyword:
        count.append(0)
        for token in tokens:
            if token in url_keyword:
                count[j] = count[j] + 1
        j = j+1
    
    j = ans =0
    max = 0
    for cnt in count:
        if(max<cnt):
            max = cnt
            ans = j
        j = j+1

    if (max==0):
        return "Nothing found"
    else:
        return urls_list[ans]
```

# give link for user query

```python
flag = True
while (flag == True):
    query = input("enter a keyword you want to search : ")
    if query == 'exit':
        break
    else:
        answer = find_link(query)
        print(answer)
```
